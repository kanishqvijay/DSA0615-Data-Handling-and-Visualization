```r
# Answer 1
library(ggplot2)
library(dplyr)

# Dataset 1: Unemployment
unemployment_data <- data.frame(
  Month = factor(c("Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"), levels = c("Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec")),
  Unemployment_Rate = c(10.7, 9.8, 10.2, 11.2, 15.75, 17.8, 19.4, 25.6, 18.6, 15.6, 26.7, 19.5)
)

ggplot(unemployment_data, aes(x = Month, y = Unemployment_Rate)) +
  geom_bar(stat = "identity", fill = "skyblue") +
  labs(title = "Unemployment Rate by Month", y = "Unemployment Rate (%)")
```

```r
# Answer 2:
library(ggplot2)
library(dplyr)

school_data <- data.frame(
  SCHOOL = rep(c("A", "B", "C", "D"), each = 3),
  GRADE_LEVEL = rep(c("Grade 1", "Grade 2", "Grade 3"), 4),
  NUMBER_OF_STUDENTS = c(25, 30, 20, 22, 28, 18, 20, 25, 15, 28, 32, 24)
)

ggplot(school_data, aes(x = GRADE_LEVEL, y = NUMBER_OF_STUDENTS, fill = SCHOOL)) +
  geom_bar(stat = "identity", position = "dodge") +
  labs(title = "Number of Students by Grade Level and School")
```

```r
# Answer 3:
library(ggplot2)
library(dplyr)

weather_data <- data.frame(
  Date = as.Date(c("2024-01-01", "2024-01-02", "2024-01-03", "2024-01-04", "2024-01-05", "2024-01-06")),
  Station_A_Temp = c(25.0, 24.5, 26.2, 23.8, 25.5, 23.0),
  Station_B_Temp = c(23.5, 22.8, 25.0, 22.5, 24.5, 21.8)
)

ggplot(weather_data) +
  geom_line(aes(x = Date, y = Station_A_Temp, color = "Station A")) +
  geom_line(aes(x = Date, y = Station_B_Temp, color = "Station B")) +
  labs(title = "Temperature Over Time", y = "Temperature (C)")
```

```r
# Answer 4
library(ggplot2)
library(dplyr)

browser_data <- data.frame(
  Year = c(2019, 2020, 2022, 2023, 2024),
  Users = c(22.7, 25.8, 28.7, 30.5, 35.2)
)

ggplot(browser_data, aes(x = Year, y = Users)) +
  geom_line() +
  geom_point() +
  labs(title = "Chrome Users Over Years", y = "Users (Millions)")
```

```r
# Answer 5
library(ggplot2)
library(dplyr)

job_data <- data.frame(
  Job_Sector = c("IT", "Government", "Customer care", "Bank", "Games"),
  Selection_Rate = c(25, 12, 45, 20, 35)
)

ggplot(job_data, aes(x = reorder(Job_Sector, -Selection_Rate), y = Selection_Rate)) +
  geom_bar(stat = "identity", fill = "lightgreen") +
  labs(title = "Selection Rate by Job Sector", x = "Job Sector", y = "Selection Rate (%)") +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))
```
